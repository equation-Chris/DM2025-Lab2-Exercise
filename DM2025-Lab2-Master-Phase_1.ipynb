## <a id='toc1_5_'></a>[**1.1 Load data**](#toc0_)

We start by loading the csv files into a single pandas dataframe for training and one for testing.

````python
# 回到 /content
%cd /content

# 有就更新；沒有就 clone
import os
if not os.path.isdir('DM2025-Lab2-Exercise/.git'):
    !git clone https://github.com/equation-Chris/DM2025-Lab2-Exercise.git
else:
    !git -C DM2025-Lab2-Exercise fetch --all --prune
    !git -C DM2025-Lab2-Exercise checkout main
    !git -C DM2025-Lab2-Exercise pull

# 檢查資料是否在
!ls -R DM2025-Lab2-Exercise/data/semeval | sed -n '1,200p'

import os, pandas as pd

BASE = "./DM2025-Lab2-Exercise/data/semeval"

def load_split(split):  # split in {"train", "dev"}
    emos = ["anger","sadness","fear","joy"]
    # choose file suffix by split
    if split == "train":
        suffix = ".train.txt"
    elif split == "dev":
        suffix = ".dev.gold.txt"
    else:
        raise ValueError("split must be 'train' or 'dev'")

    dfs = []
    for emo in emos:
        path = f"{BASE}/{split}/{emo}-ratings-0to1{suffix}"
        if not os.path.exists(path):
            raise FileNotFoundError(f"Missing file: {path}")
        df = pd.read_csv(path, sep="\t", header=None,
                         names=["id","text","emotion","intensity"])
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

# 使用
train_df = load_split("train")
test_df  = load_split("dev")

print("Train:", train_df.shape)
print("Test :", test_df.shape)

````

##### <a id='toc1_5_1_1_1_'></a>[**>>> Exercise 1 (Take home):**](#toc0_)
Plot word frequency for Top 30 words in both train and test dataset. (Hint: refer to DM lab 1)


````python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
import re
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import nltk
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# ----- 1) preprocessing / tokenization -----
STOPWORDS = set(stopwords.words('english'))

def normalize_and_tokenize(text: str):
    text = text.lower()
    # keep letters and apostrophes (e.g., don't -> don't)
    tokens = word_tokenize(text)
    tokens = [re.sub(r"[^a-z']+", "", t) for t in tokens]  # strip non-letters
    tokens = [t for t in tokens if t and t not in STOPWORDS]
    return tokens

# ----- 2) count top-N words for a dataframe (expects column 'text') -----
def top_word_counts(df: pd.DataFrame, topn=30):
    counter = Counter()
    for s in df['text'].astype(str):
        counter.update(normalize_and_tokenize(s))
    top = counter.most_common(topn)
    return pd.DataFrame(top, columns=['word', 'count'])

train_top30 = top_word_counts(train_df, topn=30)
test_top30  = top_word_counts(test_df,  topn=30)

display(train_top30.head(10))
display(test_top30.head(10))

# ----- 3) plot (horizontal bar for better readability) -----
def plot_top(df_top, title):
    plt.figure(figsize=(8, 6))
    plt.barh(df_top['word'][::-1], df_top['count'][::-1])
    plt.title(title)
    plt.xlabel('count')
    plt.tight_layout()
    plt.show()

plot_top(train_top30, 'Top-30 word frequency (TRAIN)')
plot_top(test_top30,  'Top-30 word frequency (TEST)')

# ----- 4) (optional) side-by-side comparison table -----
compare = pd.merge(train_top30, test_top30, on='word', how='outer', suffixes=('_train', '_test')).fillna(0)
compare = compare.sort_values('count_train', ascending=False).reset_index(drop=True)
display(compare.head(20))
````

````python
from sklearn.tree import DecisionTreeClassifier

# for a classificaiton problem, you need to provide both training & testing data
X_train = BOW_500.transform(train_df['text'])
y_train = train_df['emotion']

X_test = BOW_500.transform(test_df['text'])
y_test = test_df['emotion']

## take a look at data dimension is a good habit  :)
print('X_train.shape: ', X_train.shape)
print('y_train.shape: ', y_train.shape)
print('X_test.shape: ', X_test.shape)
print('y_test.shape: ', y_test.shape)
````

````python
## check by confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true=y_test, y_pred=y_test_pred)
print(cm)
````

````python
# Funciton for visualizing confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import itertools

def plot_confusion_matrix(cm, classes, title='Confusion matrix',
                          cmap=sns.cubehelix_palette(as_cmap=True)):
    """
    This function is modified from:
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    """
    classes.sort()
    tick_marks = np.arange(len(classes))

    fig, ax = plt.subplots(figsize=(5,5))
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels = classes,
           yticklabels = classes,
           title = title,
           xlabel = 'Predicted label',
           ylabel = 'True label')

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")
    ylim_top = len(classes) - 0.5
    plt.ylim([ylim_top, -.5])
    plt.tight_layout()
    plt.show()
````

````python
# plot your confusion matrix
my_tags = ['anger', 'fear', 'joy', 'sadness']
plot_confusion_matrix(cm, classes=my_tags, title='Confusion matrix')
````

Can you interpret the results above? What do they mean?



This model performed reasonably well, but we need to address a major flaw.

### 1. Accuracy Check

* **Training Accuracy (0.99):** Nearly perfect score when tested on data it already studied. This is **too high** and suggests the model might have just memorized the training examples.
* **Testing Accuracy (0.68):** The score drops significantly to 68% on new, unseen data. This **gap (0.99 vs 0.68)** confirms that our model is **overfitted**—it can't generalize its learning very well.

### 2. Performance Per Class (Classification Report)

* The **F1-Score** (a balance of Precision and Recall) is highest for **'joy' (0.72)**, meaning we are most confident when the model predicts 'joy'.
* The F1-Score is lowest for **'sadness' (0.62)**. This tells us that **'sadness' is the most difficult emotion** for the model to identify correctly and consistently.

### 3. Visualizing Errors (Confusion Matrix)

The confusion matrix shows where the model gets confused:

* **Correct Predictions (Diagonal):** The model correctly identified the most instances of **'fear' (76)**.
* **Key Errors:**
    * The model frequently mistakes **'anger'** for **'fear'** (15 times).
    * It also struggles distinguishing **'fear'** from **'sadness'** (15 instances of 'fear' predicted as 'sadness').
    * Overall, the bottom-right quadrant shows that **'sadness'** is often misclassified as other negative emotions like **'fear'**.

**Summary for the Report:**

The Decision Tree model achieved **68% testing accuracy**. While it works best for the 'joy' class, the severe overfitting shows that the current model is **not robust**. We need to apply regularization or cross-validation to improve its ability to generalize to new text data.

### <a id='toc1_10_2_'></a>[**6.2 Deal with categorical label (y)**](#toc0_)

Rather than put your label `train_df['emotion']` directly into a model, we have to process these categorical (or say nominal) label by ourselves.

Here, we use the basic method [one-hot encoding](https://en.wikipedia.org/wiki/One-hot) to transform our categorical  labels to numerical ones.


````python
# deal with label (string -> one-hot)
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_encoder.fit(y_train)

print('check label: ', label_encoder.classes_)
print('\n## Before convert')
print('y_train[0:4]:\n', y_train[0:4])
print('\ny_train.shape: ', y_train.shape)
print('y_test.shape: ', y_test.shape)

def label_encode(le, labels):
    enc = le.transform(labels)
    return keras.utils.to_categorical(enc)

def label_decode(le, one_hot_label):
    dec = np.argmax(one_hot_label, axis=1)
    return le.inverse_transform(dec)
````

````python
y_train = label_encode(label_encoder, y_train)
y_test = label_encode(label_encoder, y_test)

print('\n\n## After convert')
print('y_train[0:4]:\n', y_train[0:4])
print('\ny_train.shape: ', y_train.shape)
print('y_test.shape: ', y_test.shape)
````

### <a id='toc1_10_3_'></a>[**6.3 Build model**](#toc0_)

````python
# I/O check
input_shape = X_train.shape[1]
print('input_shape: ', input_shape)

output_shape = len(label_encoder.classes_)
print('output_shape: ', output_shape)
````

![pic2.png](https://github.com/equation-Chris/DM2025-Lab2-Exercise/blob/main/pics/pic2.png?raw=1)

````python
from keras.models import Model
from keras.layers import Input, Dense
from keras.layers import ReLU, Softmax

# input layer
model_input = Input(shape=(input_shape, ))  # 500
X = model_input

# 1st hidden layer
X_W1 = Dense(units=64)(X)  # 64
H1 = ReLU()(X_W1)

# 2nd hidden layer
H1_W2 = Dense(units=64)(H1)  # 64
H2 = ReLU()(H1_W2)

# output layer
H2_W3 = Dense(units=output_shape)(H2)  # 4
H3 = Softmax()(H2_W3)

model_output = H3

# create model
model = Model(inputs=[model_input], outputs=[model_output])

# loss function & optimizer
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# show model construction
model.summary()
````

How can we utilize these "word" vectors to represent the sentence data and train our model?


````python
# Answer here
````

The simplest way is called Averaging Word Vectors.、

1.Get Vectors: Take the Word2Vec vector (e.g., a 300-dimension number list) for every single word in the sentence

2..Add Them Up: Add all those vectors together.

3.Divide: Divide the sum by the total number of words in the sentence.

Result: This gives you one single Sentence Vector (a 300-dimension number list).This new Sentence Vector then becomes the input feature (the new $X$) for your simple models like SVM or Naive Bayes.

##### <a id='toc1_13_2_1_1_'></a>[**>>> Exercise 8 (Take home):**](#toc0_)

Generate a t-SNE and UMAP visualization to show the 15 words most related to the words "angry", "happy", "sad", "fear" (60 words total). Compare the differences between both graphs.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import umap.umap_ as umap # Need this library!

# --- 1. Prepare Target Words and Vectors (Word Embedding Data) ---
# Assuming 'w2v_google_model' (your loaded Word2Vec model) is available.

core_emotions = ["angry", "happy", "sad", "fear"]
top_n = 15 # The assignment asks for the 15 most related words per core word

all_words = []
# Create a list of all 60+ words and collect their vectors
for word in core_emotions:
    # 1. Include the core word itself
    all_words.append(word)
    try:
        # 2. Add the 15 most similar words
        similar_list = [item[0] for item in w2v_google_model.most_similar(word, topn=top_n)]
        all_words.extend(similar_list)
    except KeyError as e:
        print(f"Skipping word: {word}. Not found in model.")

# Remove any duplicates (if two core words share the same related word)
all_words = list(set(all_words))

# Extract the vectors for the final list of words
word_vectors = []
final_words = []
for word in all_words:
    try:
        word_vectors.append(w2v_google_model[word])
        final_words.append(word)
    except KeyError:
        # This will skip words that were similar to a core word but are not in the model themselves
        continue

X_vectors = np.array(word_vectors)
print(f"Total words visualized: {len(final_words)}")


# --- 2. Run Dimension Reduction ---

# t-SNE (Slower, good for local detail)
tsne_model = TSNE(n_components=2, metric='cosine', random_state=42, perplexity=15)
tsne_results = tsne_model.fit_transform(X_vectors)

# UMAP (Faster, better for global layout)
umap_model = umap.UMAP(n_components=2, metric='cosine', random_state=42, n_neighbors=15)
umap_results = umap_model.fit_transform(X_vectors)


# --- 3. Plot the Results ---

plt.figure(figsize=(15, 7))

# --- Plot 1: t-SNE ---
plt.subplot(1, 2, 1)
plt.scatter(tsne_results[:, 0], tsne_results[:, 1], s=30)
for i, word in enumerate(final_words):
    # Label each point with the word
    plt.annotate(word, (tsne_results[i, 0], tsne_results[i, 1]), textcoords="offset points", xytext=(5,5), ha='center', fontsize=8)
plt.title('t-SNE (Focus: Local Similarity)')
plt.grid(True, linestyle='--', alpha=0.6)


# --- Plot 2: UMAP ---
plt.subplot(1, 2, 2)
plt.scatter(umap_results[:, 0], umap_results[:, 1], s=30)
for i, word in enumerate(final_words):
    # Label each point with the word
    plt.annotate(word, (umap_results[i, 0], umap_results[i, 1]), textcoords="offset points", xytext=(5,5), ha='center', fontsize=8)
plt.title('UMAP (Focus: Global Structure)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
````

---

When looking at the two resulting graphs, the key differences are in what they prioritize when reducing the dimensions:

1. t-SNE (Left Graph)
Priority: Local Structure (making very similar words cluster tightly).

Appearance: The groups of related words (e.g., all "sad" words) will look extremely tight and separate from other clusters.

Main Drawback: The distance between the "happy" cluster and the "fear" cluster is often meaningless. t-SNE is bad at preserving the large-scale, global layout of the data.

2. UMAP (Right Graph)
Priority: Global Structure (preserving relationships between large clusters) while still keeping local structure.

Appearance: The groups are usually a bit looser than t-SNE's, but the negative emotion clusters (angry, sad, fear) will likely be closer to each other, and further away from the positive emotion cluster (happy).

Main Advantage: UMAP is much faster to compute, and the relative distance between the large emotion categories is more reliable than in t-SNE.

Conclusion:

For visualizing word vectors, UMAP is generally the better choice. It gives us a faster, more trustworthy picture of the overall semantic landscape by keeping both the small (local) clusters and the large (global) relationships intact.
