## 1. Model Development (10 pts Required)

In this competition I worked on emotion recognition for Tweets.  
The final system is based on fine-tuning a RoBERTa-base transformer model on the provided Twitter dataset.

### 1.1 Preprocessing Steps

**Syntax:** `###` creates a tertiary heading (H3).

**Data sources**

- `emotion.csv`: mapping from tweet `id` to the gold emotion label.
- `data_identification.csv`: mapping from tweet `id` to the split (`train` or `test`).
- `final_posts.json`: raw Twitter posts stored in a nested JSON structure.
- `samplesubmission.csv`: template file with the exact `id` ordering required by Kaggle.

**Steps**

1. **Flatten the JSON posts**  
   I used `pandas.json_normalize` on `final_posts.json` and extracted the two useful columns:
   - `root._source.post.post_id` → renamed to `id`
   - `root._source.post.text` → renamed to `text`  
   This gave me a flat DataFrame where each row is one tweet with its textual content.

2. **Merge metadata and labels**  
   - First, I merged the posts DataFrame with `data_identification.csv` on the `id` column to add the `split` information.  
   - Then, for rows with `split == "train"`, I merged with `emotion.csv` to attach the `emotion` label to each training tweet.  
   - Rows with `split == "test"` were kept as the unlabeled test set.

3. **Align test order with the submission file**  
   Kaggle expects the predictions to follow exactly the order in `samplesubmission.csv`.  
   To guarantee this, I re-merged the test posts with `samplesubmission.csv` on `id`.  
   This step ensures that `submission_roberta_simple.csv` has the same row order as the competition’s template.

4. **Basic cleaning / type handling**  
   I did not perform heavy text cleaning such as removing punctuation or lower-casing manually, because the RoBERTa tokenizer expects fairly raw text.  
   I only made sure that the `text` column is of string type before sending it to the model.

As a result of these steps, I obtained two key tables:

- `train_df(id, text, emotion)` for model training.
- `test_df(id, text)` for generating Kaggle predictions.

### 1.2 Feature Engineering Steps

Instead of traditional manual feature engineering (e.g., TF-IDF, bag-of-words, handcrafted lexicon scores), I relied on **label encoding + transformer embeddings**:

1. **Label encoding**  
   - I collected all unique emotion labels from `train_df["emotion"]`.
   - I created a mapping dictionary, e.g. `{"joy": 0, "sadness": 1, "anger": 2, "fear": 3, ...}`.
   - A new column `labels` was added to the training data by mapping each emotion string to its corresponding integer id.
   - An inverse mapping was also stored so that I can convert predictions back from ids to emotion strings when creating the submission.

2. **Train/validation split**  
   - The model input for SimpleTransformers is a DataFrame with columns `["text", "labels"]`.
   - I split this DataFrame into:
     - `train_split` (90%) and
     - `eval_split` (10%)  
     using `train_test_split` with `stratify=train_data["labels"]`.  
   - Stratification keeps the label distribution similar in both sets, which makes the validation accuracy more reliable.

3. **Implicit feature engineering by RoBERTa**  
   Although I did not manually design numerical features, RoBERTa internally converts each tweet into contextual embeddings that capture:
   - word meaning,
   - word order,
   - long-range dependencies,
   - and subtle sentiment / emotion clues.  
   This can be regarded as automatically learned feature engineering performed by the transformer.

Because the model learns features directly from raw text, no separate TF-IDF or n-gram matrices were required in this final solution.

### 1.3 Explanation of Your Model

**Model choice**

I used a **RoBERTa-base** transformer model for multi-class text classification, implemented via the `simpletransformers` library:

```text
Model type:    roberta
Pretrained ckpt: roberta-base
Task:         multi-class emotion classification
```
**Training setup**
Loss: cross-entropy over the 6 emotion classes.

Optimizer / scheduler: handled internally by SimpleTransformers (AdamW with a linear learning-rate schedule).

Hyperparameters:

num_train_epochs = 3

learning_rate = 2e-5

train_batch_size = 32

max_seq_length = 128

During training I enabled evaluate_during_training = True so the model is evaluated on eval_split periodically using accuracy.

How the model works

Each tweet is tokenized by the RoBERTa tokenizer (subword BPE tokens) and padded/truncated to 128 tokens.

The token sequence is fed into the RoBERTa encoder, which outputs contextual embeddings for each token.

The final hidden representation of the [CLS] position (or pooled output) is passed through a linear classification head to produce logits over the emotion classes.

The model is fine-tuned end-to-end on our training set so that the embeddings become specialized for Twitter emotion recognition.

Evaluation and prediction

After training, I evaluated the model on the held-out validation split using accuracy_score.
The printed value Validation Accuracy: ... reflects how well the model generalizes to unseen tweets.

For the test set, I passed the test_df["text"] list into model.predict(...) to obtain integer label ids, then converted them back to emotion strings using the inverse label map.

Finally, I created submission_roberta_simple.csv with two columns:
id and emotion (my predicted label)
ready to be uploaded to Kaggle.

### 2.2 Mention Insights You Gained

1. Transformers significantly outperform bag-of-words on this task.
Even a relatively simple RoBERTa fine-tuning setup already beats well-tuned TF-IDF + SVM models.
This shows the advantage of contextual embeddings for subtle emotional cues in short tweets.

2. Preprocessing choices matter less for RoBERTa.
With traditional models, I had to carefully decide how to normalize text, remove punctuation, handle emojis, etc.
For RoBERTa, keeping the raw text (including capitalization and punctuation) actually worked better, because the tokenizer and pretraining are designed for natural text.

````python
# 1. Install required packages (run this once in Colab)
!pip install simpletransformers pandas tqdm scikit-learn

import pandas as pd
import json
import torch
import logging
from sklearn.model_selection import train_test_split
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import sklearn

# ---------------------------------------------------------------
# Logging configuration (to see training progress)
# ---------------------------------------------------------------
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

# ---------------------------------------------------------------
# 2. Data loading and merging
# ---------------------------------------------------------------
print("--- Loading Data ---")

ID_COL = "id"
LABEL_COL = "emotion"
SUBSET_COL = "split"

# Read CSV files
emotion_df = pd.read_csv("emotion.csv")              # id, emotion
data_id_df = pd.read_csv("data_identification.csv")  # id, split (train/test)
sample_sub_df = pd.read_csv("samplesubmission.csv")  # id, emotion (dummy)

# Read JSON file with the original posts
with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

# Flatten the nested JSON structure into a DataFrame
posts_df = pd.json_normalize(posts)

# Columns specific to this dataset that contain post id and text
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"

# Keep only id + text and rename columns
posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

# Merge posts with split information (train / test)
full_df = posts_df.merge(data_id_df[[ID_COL, SUBSET_COL]],
                         how="inner", on=ID_COL)

# Training part: rows that have labels in emotion.csv
train_df = full_df[full_df[SUBSET_COL] == "train"].merge(
    emotion_df[[ID_COL, LABEL_COL]],
    how="left",
    on=ID_COL,
)

# Test part: rows without labels
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

# Align test order with samplesubmission.csv (very important for Kaggle)
test_df = sample_sub_df[[ID_COL]].merge(
    test_df[[ID_COL, "text"]],
    on=ID_COL,
    how="left",
)

print(f"Train rows: {len(train_df)}, Test rows: {len(test_df)}")
````

## 2. Feature Engineering Steps

````python
# ---------------------------------------------------------------
# 3. Prepare data format for SimpleTransformers (feature step)
# ---------------------------------------------------------------
# SimpleTransformers expects DataFrame with columns: ["text", "labels"]
# and "labels" must be integer ids (0, 1, 2, ...).

# Build a label encoder that maps emotion string -> integer id
unique_labels = train_df[LABEL_COL].unique()
label_map = {label: i for i, label in enumerate(unique_labels)}
inverse_label_map = {i: label for label, i in label_map.items()}

print("Label mapping (emotion -> id):", label_map)

# Add numeric labels column
train_df["labels"] = train_df[LABEL_COL].map(label_map)

# DataFrame used by the RoBERTa model
train_data = train_df[["text", "labels"]]

# Split into train and validation sets (90% / 10%), stratified by label
train_split, eval_split = train_test_split(
    train_data,
    test_size=0.1,
    random_state=42,
    stratify=train_data["labels"],
)

# Make sure the text column is string type
train_split["text"] = train_split["text"].astype(str)
eval_split["text"] = eval_split["text"].astype(str)
````

## 3. Model Implementation Steps

````python
# ---------------------------------------------------------------
# 4. RoBERTa configuration (model hyperparameters)
# ---------------------------------------------------------------
model_args = ClassificationArgs()

# Main training hyperparameters
model_args.num_train_epochs = 3           # number of training epochs
model_args.learning_rate = 2e-5          # standard fine-tuning learning rate
model_args.train_batch_size = 32         # adjust depending on GPU memory (16/32)
model_args.eval_batch_size = 32
model_args.max_seq_length = 128          # tweets are usually short; 128 is enough

# Misc settings
model_args.overwrite_output_dir = True
model_args.save_steps = -1               # do not save intermediate checkpoints
model_args.save_model_every_epoch = False
model_args.evaluate_during_training = True
model_args.evaluate_during_training_verbose = True
model_args.use_multiprocessing = False
model_args.use_multiprocessing_for_evaluation = False

# Initialize a RoBERTa-base model for multi-class classification
model = ClassificationModel(
    "roberta",
    "roberta-base",
    num_labels=len(unique_labels),
    args=model_args,
    use_cuda=torch.cuda.is_available(),
)

# ---------------------------------------------------------------
# 5. Training
# ---------------------------------------------------------------
print("\n--- Starting Training (RoBERTa) ---")

# Train the model, evaluating on the validation split with accuracy metric
model.train_model(
    train_split,
    eval_df=eval_split,
    acc=sklearn.metrics.accuracy_score,
)

# ---------------------------------------------------------------
# 6. Evaluation on validation set
# ---------------------------------------------------------------
print("\n--- Evaluating ---")

result, model_outputs, wrong_predictions = model.eval_model(
    eval_split,
    acc=sklearn.metrics.accuracy_score,
)

print(f"Validation Accuracy: {result['acc']:.4f}")

# ---------------------------------------------------------------
# 7. Predict on test set and create Kaggle submission
# ---------------------------------------------------------------
print("\n--- Predicting Test Set ---")

# Texts to predict
to_predict = test_df["text"].tolist()

# Get predictions (as integer label ids)
predictions, raw_outputs = model.predict(to_predict)

# Convert integer labels back to original emotion strings
final_preds_labels = [inverse_label_map[pred] for pred in predictions]

# Build submission file
submission = sample_sub_df.copy()
submission[LABEL_COL] = final_preds_labels
submission.to_csv("submission_roberta_simple.csv", index=False)

print("Saved submission_roberta_simple.csv")
submission.head()
````
