# **Competition Code**

## 1. Preprocessing Steps

```python
### Preprocessing: load data, flatten JSON, merge train/test, and clean text

import pandas as pd
import json
import re

# 1. Column name constants (match your CSVs)
ID_COL = "id"          # id column in emotion.csv and data_identification.csv
LABEL_COL = "emotion"  # label column in emotion.csv
SUBSET_COL = "split"   # split column in data_identification.csv ("train"/"test")

# 2. Load CSV files
emotion_df = pd.read_csv("emotion.csv")
data_id_df = pd.read_csv("data_identification.csv")
sample_sub_df = pd.read_csv("samplesubmission.csv")

print("emotion_df:", emotion_df.shape, "columns:", emotion_df.columns.tolist())
print("data_id_df:", data_id_df.shape, "columns:", data_id_df.columns.tolist())
print("sample_sub_df:", sample_sub_df.shape, "columns:", sample_sub_df.columns.tolist())

# 3. Load JSON file with posts and flatten it
with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

posts_df = pd.json_normalize(posts)
print("posts_df (raw) shape:", posts_df.shape)

# Choose correct text/id columns
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"

# Keep only id + text and rename to match CSV id column
posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

print("posts_df (id + text) shape:", posts_df.shape)
print(posts_df.head())

# 4. Merge posts with split information (train/test)
full_df = posts_df.merge(
    data_id_df[[ID_COL, SUBSET_COL]],
    how="inner",
    on=ID_COL,
)

# Train part: rows with split == "train", then merge labels
train_df = full_df[full_df[SUBSET_COL] == "train"].merge(
    emotion_df[[ID_COL, LABEL_COL]],
    how="left",
    on=ID_COL,
)

# Test part: rows with split == "test" (no labels)
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

print("train_df shape:", train_df.shape)
print("test_df shape:", test_df.shape)
print(train_df[[ID_COL, "text", LABEL_COL]].head())

# 5. Text cleaning function
def clean_text(text: str) -> str:
    """Basic text normalization for tweets."""
    text = str(text).lower()                     # lowercase
    text = re.sub(r"http\S+", " ", text)        # remove URLs
    text = re.sub(r"@\w+", " ", text)           # remove @mentions
    text = re.sub(r"#", " ", text)              # remove '#' but keep hashtag word
    text = re.sub(r"[^a-z\s]", " ", text)       # keep only letters and spaces
    text = re.sub(r"\s+", " ", text).strip()    # collapse multiple spaces
    return text

# Apply cleaning to train and test texts
train_df["clean_text"] = train_df["text"].apply(clean_text)
test_df["clean_text"] = test_df["text"].apply(clean_text)

print("Columns in train_df:", train_df.columns.tolist())
print("Columns in test_df:", test_df.columns.tolist())
print(train_df[["text", "clean_text", LABEL_COL]].head())
print(test_df[["text", "clean_text"]].head())
```

## 2. Feature Engineering Steps

```python
### Feature Engineering: combine word- and char-level TF-IDF features

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

# 1. Extract cleaned texts and labels from train_df
X_text = train_df["clean_text"].values          # preprocessed tweets
y = train_df["emotion"].values                  # target labels

# 2. Split into training and validation sets (hold-out)
X_train_text, X_valid_text, y_train, y_valid = train_test_split(
    X_text,
    y,
    test_size=0.15,      # 15% for validation, 85% for training
    random_state=42,
    stratify=y
)

# 3. Word-level TF-IDF (unigrams + bigrams)
word_tfidf = TfidfVectorizer(
    analyzer="word",
    ngram_range=(1, 2),
    max_features=40000,
    min_df=3,
    max_df=0.90,
    sublinear_tf=True
)

X_train_word = word_tfidf.fit_transform(X_train_text)
X_valid_word = word_tfidf.transform(X_valid_text)
X_test_word  = word_tfidf.transform(test_df["clean_text"].values)

# 4. Character-level TF-IDF (3- to 5-gram characters)
char_tfidf = TfidfVectorizer(
    analyzer="char",
    ngram_range=(3, 5),
    max_features=40000,
    min_df=5,
    sublinear_tf=True
)

X_train_char = char_tfidf.fit_transform(X_train_text)
X_valid_char = char_tfidf.transform(X_valid_text)
X_test_char  = char_tfidf.transform(test_df["clean_text"].values)

# 5. Concatenate word and char features horizontally
X_train = hstack([X_train_word, X_train_char])
X_valid = hstack([X_valid_word, X_valid_char])
X_test  = hstack([X_test_word,  X_test_char])

print("X_train shape:", X_train.shape)
print("X_valid shape:", X_valid.shape)
print("X_test  shape:", X_test.shape)
```

## 3. Model Implementation Steps

```python
### Model: tune LinearSVC, evaluate on validation, and create submission

from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from scipy.sparse import vstack, hstack
import numpy as np

# 1. Define base classifier and hyperparameter grid
base_clf = LinearSVC(random_state=42)

param_grid = {
    "C": [0.25, 0.5, 1.0, 2.0],          # regularization strength
    "class_weight": [None, "balanced"]   # handle class imbalance or not
}

# 2. Set up cross-validation on the training split
cv = StratifiedKFold(
    n_splits=3,
    shuffle=True,
    random_state=42
)

grid = GridSearchCV(
    estimator=base_clf,
    param_grid=param_grid,
    scoring="f1_macro",   # optimize macro F1
    cv=cv,
    n_jobs=-1,
    verbose=1
)

# 3. Run grid search on the training split (X_train, y_train)
grid.fit(X_train, y_train)

print("Best parameters from CV:", grid.best_params_)
print(f"Best CV F1 (macro): {grid.best_score_:.4f}")

# 4. Evaluate the best model on the validation set (hold-out)
best_clf = grid.best_estimator_

y_valid_pred = best_clf.predict(X_valid)

val_acc = accuracy_score(y_valid, y_valid_pred)
val_f1_macro = f1_score(y_valid, y_valid_pred, average="macro")

print(f"Validation accuracy (hold-out): {val_acc:.4f}")
print(f"Validation F1 (macro, hold-out): {val_f1_macro:.4f}")

# 5. Retrain a final model on ALL labeled data (train + valid)

X_full = vstack([X_train, X_valid])
y_full = np.concatenate([y_train, y_valid])

final_clf = LinearSVC(
    C=best_clf.C,
    class_weight=best_clf.class_weight,
    random_state=42
)

final_clf.fit(X_full, y_full)

# 6. Prepare test set in the same id order as sample_sub_df
test_for_pred = sample_sub_df[["id"]].merge(
    test_df[["id", "clean_text"]],
    on="id",
    how="left"
)

# Build word + char features for the test set
X_test_word  = word_tfidf.transform(test_for_pred["clean_text"].values)
X_test_char  = char_tfidf.transform(test_for_pred["clean_text"].values)
X_test_all   = hstack([X_test_word, X_test_char])

# 7. Predict labels for the test set using the final model
test_pred = final_clf.predict(X_test_all)

# 8. Fill predictions into the sample submission format
submission = sample_sub_df.copy()    # columns: ['id', 'emotion']
submission["emotion"] = test_pred    # overwrite the emotion column with our predictions

# 9. Save the submission file
submission.to_csv("submission.csv", index=False)
submission.head()
```

```python
### Preprocessing: load data, flatten JSON, merge train/test, and clean text
import pandas as pd
import json
import re
import emoji
import numpy as np
from nltk.tokenize import TweetTokenizer
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack, vstack
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import ComplementNB
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, f1_score

# 1. Column name constants
ID_COL = "id"
LABEL_COL = "emotion"
SUBSET_COL = "split"

# 2. Load CSV files
emotion_df = pd.read_csv("emotion.csv")
data_id_df = pd.read_csv("data_identification.csv")
sample_sub_df = pd.read_csv("samplesubmission.csv")

print("--- Data Loading ---")
print("emotion_df:", emotion_df.shape, "columns:", emotion_df.columns.tolist())
print("data_id_df:", data_id_df.shape, "columns:", data_id_df.columns.tolist())
print("sample_sub_df:", sample_sub_df.shape, "columns:", sample_sub_df.columns.tolist())

# 3. Load JSON file with posts and flatten it
with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

posts_df = pd.json_normalize(posts)
print("posts_df (raw) shape:", posts_df.shape)

# Choose correct text/id columns
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"

# Keep only id + text and rename to match CSV id column
posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

print("posts_df (id + text) shape:", posts_df.shape)
print(posts_df.head())

# 4. Merge posts with split information (train/test)
full_df = posts_df.merge(
    data_id_df[[ID_COL, SUBSET_COL]],
    how="inner",
    on=ID_COL,
)

# Train part: rows with split == "train", then merge labels
train_df = full_df[full_df[SUBSET_COL] == "train"].merge(
    emotion_df[[ID_COL, LABEL_COL]],
    how="left",
    on=ID_COL,
)

# Test part: rows with split == "test" (no labels)
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

print("\n--- Data Splitting ---")
print("train_df shape:", train_df.shape)
print("test_df shape:", test_df.shape)
print(train_df[[ID_COL, "text", LABEL_COL]].head())

# 5. Optimized Text cleaning function (Preserve Emojis and tone)
def clean_text_optimized(text: str) -> str:
    text = str(text)
    # Convert Emojis to text description (e.g., üò≠ -> :loudly_crying_face:)
    text = emoji.demojize(text, delimiters=(" ", " "))
    text = text.lower()
    text = re.sub(r"http\S+", " ", text)  # remove URLs
    text = re.sub(r"@\w+", " ", text)     # remove @mentions
    text = re.sub(r"#", " ", text)        # remove hash sign
    # Keep letters, numbers, colons (for emojis), exclamation marks, question marks
    text = re.sub(r"[^a-z0-9\s:!?_]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Apply cleaning
train_df["clean_text"] = train_df["text"].apply(clean_text_optimized)
test_df["clean_text"] = test_df["text"].apply(clean_text_optimized)

print("\n--- Text Cleaning (Comparison) ---")
print(train_df[["text", "clean_text", LABEL_COL]].head())
print(test_df[["text", "clean_text"]].head())

### Feature Engineering: combine word- and char-level TF-IDF features

# 1. Extract cleaned texts and labels from train_df
X_text = train_df["clean_text"].values
y = train_df["emotion"].values

# 2. Split into training and validation sets (hold-out) for evaluation metrics
X_train_text, X_valid_text, y_train, y_valid = train_test_split(
    X_text,
    y,
    test_size=0.15,      # 15% for validation
    random_state=42,
    stratify=y
)

# 3. Optimized Tokenizer
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
def my_tokenizer(text):
    return tokenizer.tokenize(text)

print("\n--- Feature Engineering ---")
# 4. Word-level TF-IDF (Enhanced)
word_tfidf = TfidfVectorizer(
    tokenizer=my_tokenizer,
    ngram_range=(1, 3),      # 1-3 grams
    max_features=50000,
    min_df=2,
    sublinear_tf=True
)

print("Fitting Word TF-IDF...")
X_train_word = word_tfidf.fit_transform(X_train_text)
X_valid_word = word_tfidf.transform(X_valid_text)
X_test_word  = word_tfidf.transform(test_df["clean_text"].values)

# 5. Character-level TF-IDF (Enhanced)
char_tfidf = TfidfVectorizer(
    analyzer="char",
    ngram_range=(3, 6),      # 3-6 grams
    max_features=30000,
    sublinear_tf=True
)

print("Fitting Char TF-IDF...")
X_train_char = char_tfidf.fit_transform(X_train_text)
X_valid_char = char_tfidf.transform(X_valid_text)
X_test_char  = char_tfidf.transform(test_df["clean_text"].values)

# 6. Concatenate
X_train = hstack([X_train_word, X_train_char])
X_valid = hstack([X_valid_word, X_valid_char])
X_test  = hstack([X_test_word,  X_test_char])

print("X_train shape:", X_train.shape)
print("X_valid shape:", X_valid.shape)
print("X_test  shape:", X_test.shape)

### Model: Ensemble Voting Classifier

print("\n--- Model Training & Evaluation ---")

# 1. Define base classifiers
clf1 = LinearSVC(C=1.0, class_weight="balanced", random_state=42, max_iter=10000)
clf2 = LogisticRegression(C=2.0, class_weight="balanced", solver='liblinear', random_state=42, max_iter=1000)
clf3 = ComplementNB(alpha=0.5)

# 2. Ensemble
voting_clf = VotingClassifier(
    estimators=[('svc', clf1), ('lr', clf2), ('nb', clf3)],
    voting='hard'
)

# 3. Cross-Validation (Simulating GridSearch Best Score)
# Calculate CV score here to replace the original GridSearch "Best CV F1"
cv_scores = cross_val_score(voting_clf, X_train, y_train, cv=3, scoring="f1_macro", n_jobs=-1)
print(f"Ensemble Cross-Validation F1 (macro): {cv_scores.mean():.4f}")

# 4. Train on split-train and Evaluate on split-valid
voting_clf.fit(X_train, y_train)
y_valid_pred = voting_clf.predict(X_valid)

val_acc = accuracy_score(y_valid, y_valid_pred)
val_f1_macro = f1_score(y_valid, y_valid_pred, average="macro")

print(f"Validation accuracy (hold-out): {val_acc:.4f}")
print(f"Validation F1 (macro, hold-out): {val_f1_macro:.4f}")

# 5. Retrain on ALL labeled data (train + valid)
print("\n--- Retraining on Full Data ---")
X_full = vstack([X_train, X_valid])
y_full = np.concatenate([y_train, y_valid])

final_clf = VotingClassifier(
    estimators=[('svc', clf1), ('lr', clf2), ('nb', clf3)],
    voting='hard'
)

final_clf.fit(X_full, y_full)
print("Final model trained.")

# 6. Predict on Test
# Prepare test set in the same id order as sample_sub_df
test_for_pred = sample_sub_df[["id"]].merge(
    test_df[["id", "clean_text"]],
    on="id",
    how="left"
)

# Build features for test set
X_test_word_final = word_tfidf.transform(test_for_pred["clean_text"].values)
X_test_char_final = char_tfidf.transform(test_for_pred["clean_text"].values)
X_test_all = hstack([X_test_word_final, X_test_char_final])

# 7. Predict
test_pred = final_clf.predict(X_test_all)

# 8. Submission
submission = sample_sub_df.copy()
submission["emotion"] = test_pred
submission.to_csv("submission_optimized.csv", index=False)

print("\n--- Submission Generated ---")
print(submission.head())
print("Saved to submission_optimized.csv")
```

```python
### Preprocessing: load data, flatten JSON, merge train/test, and clean text
import pandas as pd
import json
import re
import emoji
import numpy as np
from nltk.tokenize import TweetTokenizer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack, vstack
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import accuracy_score, f1_score

# 1. ÂÆöÁæ©Ê¨Ñ‰ΩçÂêçÁ®±Â∏∏Êï∏
ID_COL = "id"
LABEL_COL = "emotion"
SUBSET_COL = "split"

# 2. ËÆÄÂèñ CSV Ê™îÊ°à
print("--- Data Loading ---")
emotion_df = pd.read_csv("emotion.csv")
data_id_df = pd.read_csv("data_identification.csv")
sample_sub_df = pd.read_csv("samplesubmission.csv")

# 3. ËÆÄÂèñ JSON ‰∏¶Â±ïÂπ≥
with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

posts_df = pd.json_normalize(posts)

# ÈÅ∏ÊìáÊ≠£Á¢∫ÁöÑÊ¨Ñ‰Ωç
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"

posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

# 4. Âêà‰ΩµË≥áÊñôËàáÂàÜÂâ≤Ë≥áË®ä
full_df = posts_df.merge(data_id_df[[ID_COL, SUBSET_COL]], how="inner", on=ID_COL)

# Ë®ìÁ∑¥ÈõÜÈÉ®ÂàÜ
train_df = full_df[full_df[SUBSET_COL] == "train"].merge(
    emotion_df[[ID_COL, LABEL_COL]], how="left", on=ID_COL
)

# Ê∏¨Ë©¶ÈõÜÈÉ®ÂàÜ
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# 5. ÂÑ™ÂåñÁâàÊñáÂ≠óÊ∏ÖÊ¥ó (‰øùÁïô Emoji)
def clean_text_optimized(text: str) -> str:
    text = str(text)
    # Â∞á Emoji ËΩâÁÇ∫ÊñáÂ≠óÊèèËø∞ (e.g., üò≠ -> :loudly_crying_face:)
    text = emoji.demojize(text, delimiters=(" ", " "))
    text = text.lower()
    text = re.sub(r"http\S+", " ", text)
    text = re.sub(r"@\w+", " ", text)
    text = re.sub(r"#", " ", text)
    # ‰øùÁïôÂ≠óÊØç„ÄÅÊï∏Â≠ó„ÄÅ‰ª•Âèä emoji ËΩâÊèõÂæåÁöÑÂÜíËôüËàáÁâπÂÆöÊ®ôÈªû
    text = re.sub(r"[^a-z0-9\s:!?_']", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

train_df["clean_text"] = train_df["text"].apply(clean_text_optimized)
test_df["clean_text"] = test_df["text"].apply(clean_text_optimized)

### ÁâπÂæµÂ∑•Á®ã (Feature Engineering)
print("\n--- Feature Engineering ---")

X_text = train_df["clean_text"].values
y = train_df["emotion"].values

# ÂàáÂàÜÈ©óË≠âÈõÜ (Validation Split)
X_train_text, X_valid_text, y_train, y_valid = train_test_split(
    X_text, y, test_size=0.1, random_state=42, stratify=y
)

# ‰ΩøÁî® NLTK ÁöÑ TweetTokenizer
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
def my_tokenizer(text):
    return tokenizer.tokenize(text)

# Word-level TF-IDF
word_tfidf = TfidfVectorizer(
    tokenizer=my_tokenizer,
    ngram_range=(1, 4),      # Â¢ûÂä†Âà∞ 4-grams ‰ª•ÊçïÊçâÊõ¥Â§ö‰∏ä‰∏ãÊñá
    max_features=60000,      # Â¢ûÂä†ÁâπÂæµÊï∏Èáè
    min_df=2,
    sublinear_tf=True
)

# Char-level TF-IDF
char_tfidf = TfidfVectorizer(
    analyzer="char",
    ngram_range=(3, 6),
    max_features=40000,
    sublinear_tf=True
)

print("Vectorizing...")
X_train_word = word_tfidf.fit_transform(X_train_text)
X_valid_word = word_tfidf.transform(X_valid_text)
X_test_word  = word_tfidf.transform(test_df["clean_text"].values)

X_train_char = char_tfidf.fit_transform(X_train_text)
X_valid_char = char_tfidf.transform(X_valid_text)
X_test_char  = char_tfidf.transform(test_df["clean_text"].values)

# Âêà‰ΩµÁâπÂæµ
X_train = hstack([X_train_word, X_train_char])
X_valid = hstack([X_valid_word, X_valid_char])
X_test  = hstack([X_test_word,  X_test_char])

### Ê®°ÂûãË®ìÁ∑¥ÔºöÈáùÂ∞ç Accuracy ÂÑ™ÂåñÁöÑÈõÜÊàêÊ®°Âûã
print("\n--- Model Training (Optimized for Accuracy) ---")

# 1. Logistic Regression
# ÈóúÈçµÊîπÂãïÔºöclass_weight=None (ÁßªÈô§Âπ≥Ë°°Ê¨äÈáçÔºåËÆìÊ®°ÂûãÂ∞àÊ≥®ÊñºÊ∫ñÁ¢∫Áéá)
clf_lr = LogisticRegression(
    C=4.0,
    solver='liblinear',
    max_iter=2000,
    class_weight=None,
    random_state=42
)

# 2. LinearSVC (Á∂ìÈÅéÊ†°Ê∫ñ‰ª•Ëº∏Âá∫Ê©üÁéá)
# LinearSVC ÈÄöÂ∏∏ÊØî LR Ê∫ñÔºå‰ΩÜÈúÄË¶Å Calibration ÊâçËÉΩÁî®Êñº Soft Voting
svc_base = LinearSVC(C=1.0, class_weight=None, max_iter=5000, random_state=42)
clf_svc = CalibratedClassifierCV(svc_base, method='sigmoid', cv=3)

# 3. RidgeClassifier (Á∂ìÈÅéÊ†°Ê∫ñ)
# Â∞çÊñáÂ≠óÂàÜÈ°ûÈùûÂ∏∏Á©©ÂÅ•
ridge_base = RidgeClassifier(alpha=1.0, class_weight=None, random_state=42)
clf_ridge = CalibratedClassifierCV(ridge_base, method='sigmoid', cv=3)

# 4. MultinomialNB (Â¢ûÂä†Â§öÊ®£ÊÄß)
clf_nb = MultinomialNB(alpha=0.1)

# ‰ΩøÁî® Soft Voting ÈõÜÊàê
# Soft voting (‰ΩøÁî®Ê©üÁéáÂπ≥Âùá) ÈÄöÂ∏∏ÊØî Hard voting (Êï∏‰∫∫È†≠) Ê∫ñÁ¢∫ÁéáÊõ¥È´ò
voting_clf = VotingClassifier(
    estimators=[
        ('lr', clf_lr),
        ('svc', clf_svc),
        ('ridge', clf_ridge),
        ('nb', clf_nb)
    ],
    voting='soft',
    weights=[2, 2, 2, 1] # Áµ¶‰∫àÂº∑ÊïàÁöÑÁ∑öÊÄßÊ®°ÂûãÊõ¥È´òÁöÑÊ¨äÈáç
)

# Ë®ìÁ∑¥
voting_clf.fit(X_train, y_train)

# È©óË≠â
y_valid_pred = voting_clf.predict(X_valid)

val_acc = accuracy_score(y_valid, y_valid_pred)
val_f1 = f1_score(y_valid, y_valid_pred, average="macro")

print(f"Validation Accuracy: {val_acc:.4f}")
print(f"Validation F1 Macro: {val_f1:.4f}")

if val_acc >= 0.7:
    print(">>> SUCCESS: ÈÅîÊàê Accuracy >= 0.7ÔºÅ")
else:
    print(f">>> ÁõÆÂâç Accuracy ÁÇ∫ {val_acc:.4f}„ÄÇ")

### ÊúÄÁµÇÔºö‰ΩøÁî®ÂÆåÊï¥Ë≥áÊñôÈáçÊñ∞Ë®ìÁ∑¥‰∏¶Áî¢ÁîüÊèê‰∫§Ê™î
print("\n--- Retraining on Full Data ---")
X_full = vstack([X_train, X_valid])
y_full = np.concatenate([y_train, y_valid])

voting_clf.fit(X_full, y_full)

# È†êÊ∏¨Ê∏¨Ë©¶ÈõÜ
test_pred = voting_clf.predict(X_test)

# ÂÑ≤Â≠òÁµêÊûú
submission = sample_sub_df.copy()
submission["emotion"] = test_pred
submission.to_csv("submission_accuracy_optimized.csv", index=False)
print("Saved to submission_accuracy_optimized.csv")
```

```python
### Preprocessing: load data, flatten JSON, merge train/test, and clean text
import pandas as pd
import json
import re
import emoji
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression, SGDClassifier, RidgeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import accuracy_score, f1_score
from scipy.sparse import hstack, vstack

# 1. Column name constants
ID_COL = "id"
LABEL_COL = "emotion"
SUBSET_COL = "split"

# 2. Load Data
print("--- Data Loading ---")
emotion_df = pd.read_csv("emotion.csv")
data_id_df = pd.read_csv("data_identification.csv")
sample_sub_df = pd.read_csv("samplesubmission.csv")

with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

posts_df = pd.json_normalize(posts)
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"
posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

# Merge
full_df = posts_df.merge(data_id_df[[ID_COL, SUBSET_COL]], how="inner", on=ID_COL)
train_df = full_df[full_df[SUBSET_COL] == "train"].merge(emotion_df[[ID_COL, LABEL_COL]], how="left", on=ID_COL)
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

print(f"Train rows: {len(train_df)}, Test rows: {len(test_df)}")

# 3. Minimal Cleaning (‰øùÁïôÊúÄÂ§öË≥áË®ä)
def clean_text_final(text: str) -> str:
    text = str(text)
    # ËΩâÊèõ emoji ‰ΩÜ‰∏çÂä†Á©∫Ê†ºÔºå‰øùÁïôÂÉèÊòØ :smile: ÈÄôÊ®£ÁöÑÁâπÂæµ
    text = emoji.demojize(text)
    # Âè™ÁßªÈô§Á∂≤ÂùÄÂíå@Ôºå‰øùÁïô #hashtag ÁöÑÊñáÂ≠ó
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    # ‰∏çËΩâÂ∞èÂØ´ÔºÅ(Case matters for anger) -> ‰ΩÜ TF-IDF ÊúÉËá™Â∑±ËΩâ
    # ÈÄôË£°ÊàëÂÄëÂè™ÂõûÂÇ≥ÂéüÂßãÂ≠ó‰∏≤ÔºåËÆìÂæåÈù¢ÁöÑ HandCraftedFeature ÊäìÂ§ßÂØ´
    return text.strip()

# 4. ÊâãÂ∑•ÁâπÂæµÊèêÂèñÂô® (ÊçïÊçâ È©öÂòÜËôü„ÄÅÂïèËôü„ÄÅÂ§ßÂØ´Á®ãÂ∫¶)
class HandCraftedFeatures(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        features = []
        for text in X:
            text_len = len(text)
            # Ë®àÁÆóÂ§ßÂØ´Â≠óÊØçÊØî‰æã (ÊÜ§ÊÄí/ÊøÄÂãï)
            caps_count = sum(1 for c in text if c.isupper())
            caps_ratio = caps_count / (text_len + 1)
            # Ë®àÁÆóÊ®ôÈªûÁ¨¶Ëôü (È©öÂñú/ÁñëÂïè)
            excl_count = text.count('!')
            ques_count = text.count('?')
            # È°èÊñáÂ≠óÂà§Êñ∑ (Á∞°ÂñÆÁâà)
            has_smile = 1 if ':)' in text or ':-)' in text else 0
            has_sad = 1 if ':(' in text or ':-(' in text else 0

            features.append([text_len, caps_count, caps_ratio, excl_count, ques_count, has_smile, has_sad])
        return np.array(features)

# Ê∫ñÂÇôË≥áÊñô
# Ê≥®ÊÑèÔºöTF-IDF ÈúÄË¶ÅÂ∞èÂØ´ÔºåÊâãÂ∑•ÁâπÂæµÈúÄË¶ÅÂéüÂßãÂ§ßÂ∞èÂØ´
# ÊàëÂÄëÂÖàÂÅö‰∏ÄÂÄãÁµ±‰∏ÄÁöÑ list
raw_train_text = train_df["text"].apply(clean_text_final).values
raw_test_text = test_df["text"].apply(clean_text_final).values
y = train_df["emotion"].values

# ÂàáÂàÜÈ©óË≠âÈõÜ (Train 90%, Valid 10%)
X_train_raw, X_valid_raw, y_train, y_valid = train_test_split(
    raw_train_text, y, test_size=0.1, random_state=42, stratify=y
)

print("\n--- Feature Engineering (Stacking Strategy) ---")

# A. TF-IDF Pipeline (Ëá™ÂãïËΩâÂ∞èÂØ´)
# Word-level: ÁØÑÂúçÈñãÂà∞ 1-4 gramsÔºå‰∏çË®≠ min_df ‰ª•‰øùÁïôÊâÄÊúâÁ®ÄÊúâË©û
tfidf_word = TfidfVectorizer(
    analyzer='word',
    ngram_range=(1, 4),
    max_features=80000,
    min_df=1,
    sublinear_tf=True,
    strip_accents='unicode'
)

# Char-level: ÊçïÊçâÊãºÂ≠óËÆäÂåñ
tfidf_char = TfidfVectorizer(
    analyzer='char',
    ngram_range=(2, 6),
    max_features=50000,
    sublinear_tf=True
)

# B. Hand Crafted Pipeline
hand_feats = Pipeline([
    ('extractor', HandCraftedFeatures()),
    ('scaler', MinMaxScaler()) # Ê≠∏‰∏ÄÂåñÂæàÈáçË¶Å
])

print("Vectorizing Word Features...")
X_train_word = tfidf_word.fit_transform(X_train_raw)
X_valid_word = tfidf_word.transform(X_valid_raw)
X_test_word = tfidf_word.transform(raw_test_text)

print("Vectorizing Char Features...")
X_train_char = tfidf_char.fit_transform(X_train_raw)
X_valid_char = tfidf_char.transform(X_valid_raw)
X_test_char = tfidf_char.transform(raw_test_text)

print("Extracting Hand-crafted Features...")
X_train_hand = hand_feats.fit_transform(X_train_raw)
X_valid_hand = hand_feats.transform(X_valid_raw)
X_test_hand = hand_feats.transform(raw_test_text)

# Âêà‰ΩµÊâÄÊúâÁâπÂæµ (Sparse + Dense)
X_train = hstack([X_train_word, X_train_char, X_train_hand])
X_valid = hstack([X_valid_word, X_valid_char, X_valid_hand])
X_test = hstack([X_test_word, X_test_char, X_test_hand])

print(f"Final Feature Matrix Shape: {X_train.shape}")

# 5. Model: Stacking Classifier
# ÈÄôÊòØÊØî Voting Êõ¥Âº∑ÁöÑÊäÄË°ìÔºå‰ΩøÁî® Meta-Learner ‰øÆÊ≠£ÈåØË™§
print("\n--- Training Stacking Classifier ---")

# Level 1 Models (Âü∫Ê®°Âûã)
estimators = [
    ('svm', CalibratedClassifierCV(LinearSVC(C=0.5, random_state=42, max_iter=3000))), # Èôç‰Ωé C Èò≤Ê≠¢ÈÅéÊì¨Âêà
    ('lr', LogisticRegression(C=2.0, solver='liblinear', random_state=42)),
    ('sgd', SGDClassifier(loss='log_loss', alpha=1e-4, random_state=42)), # Âø´ÈÄü‰∏îÂº∑Â§ß
    ('mnb', MultinomialNB(alpha=0.1)) # Â¢ûÂä†Â§öÊ®£ÊÄß
]

# Level 2 Meta-Learner (Êï¥ÂêàÊ®°Âûã)
# ‰ΩøÁî® Logistic Regression ‰æÜÂàÜÈÖçÂêÑÂÄãÂü∫Ê®°ÂûãÁöÑÊ¨äÈáç
final_estimator = LogisticRegression(C=1.0, random_state=42)

clf = StackingClassifier(
    estimators=estimators,
    final_estimator=final_estimator,
    cv=3, # ÂÖßÈÉ®‰∫§ÂèâÈ©óË≠â
    n_jobs=-1,
    passthrough=False # Âè™‰ΩøÁî®Âü∫Ê®°ÂûãÁöÑÈ†êÊ∏¨ÁµêÊûú‰ΩúÁÇ∫ÁâπÂæµ
)

# Ë®ìÁ∑¥
clf.fit(X_train, y_train)

# È©óË≠â
print("Evaluating on Validation Set...")
y_pred_valid = clf.predict(X_valid)
acc = accuracy_score(y_valid, y_pred_valid)
f1 = f1_score(y_valid, y_pred_valid, average='macro')

print(f"Validation Accuracy: {acc:.4f}")
print(f"Validation F1 (Macro): {f1:.4f}")

# 6. Retrain Full & Predict
if acc > 0.63: # Âè™Ë¶ÅÊØî‰πãÂâçÂ•ΩÂ∞±ÈáçÊñ∞Ë®ìÁ∑¥
    print("\n--- Retraining on Full Data & Predicting ---")
    # Âêà‰Ωµ Train + Valid
    X_full = vstack([X_train, X_valid])
    y_full = np.concatenate([y_train, y_valid])

    clf.fit(X_full, y_full)

    test_preds = clf.predict(X_test)

    submission = sample_sub_df.copy()
    submission["emotion"] = test_preds
    submission.to_csv("submission_stacking.csv", index=False)
    print("Saved to submission_stacking.csv")
else:
    print("Score did not improve significantly. Check data quality.")
```

```python
### Competition Code: Hybrid TF-IDF + Hand-crafted Features + Ensemble

import pandas as pd
import json
import re
import emoji
import numpy as np

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV

from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import accuracy_score, f1_score
from scipy.sparse import hstack, vstack

# -------------------------------------------------------------------
# 1. Load and merge data
# -------------------------------------------------------------------
ID_COL = "id"
LABEL_COL = "emotion"
SUBSET_COL = "split"

emotion_df = pd.read_csv("emotion.csv")
data_id_df = pd.read_csv("data_identification.csv")
sample_sub_df = pd.read_csv("samplesubmission.csv")

with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

posts_df = pd.json_normalize(posts)
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"

posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

# Merge with split information
full_df = posts_df.merge(data_id_df[[ID_COL, SUBSET_COL]], how="inner", on=ID_COL)

train_df = full_df[full_df[SUBSET_COL] == "train"].merge(
    emotion_df[[ID_COL, LABEL_COL]], how="left", on=ID_COL
)
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

print(f"Train rows: {len(train_df)}, Test rows: {len(test_df)}")

# -------------------------------------------------------------------
# 2. Text cleaning and hand-crafted features
# -------------------------------------------------------------------
def clean_text_final(text: str) -> str:
    """Advanced text cleaning: keep emojis (demojized), remove URLs/mentions, expand contractions."""
    text = str(text)
    text = emoji.demojize(text)                     # keep emoji information as tokens like :smile:
    text = re.sub(r"http\S+", "", text)             # remove URLs
    text = re.sub(r"@\w+", "", text)                # remove @mentions
    text = re.sub(r"n\'t", " not", text)            # simple contraction handling
    text = re.sub(r"\'re", " are", text)
    return text.strip()

class AdvancedFeatures(BaseEstimator, TransformerMixin):
    """Extract structural + tiny sentiment-lexicon features."""
    def __init__(self):
        self.anger_words = set(['hate', 'stupid', 'fuck', 'shit', 'angry', 'mad', 'hell', 'kill', 'shut', 'damn', 'rage', 'annoying'])
        self.joy_words   = set(['love', 'happy', 'great', 'good', 'best', 'awesome', 'smile', 'fun', 'excited', 'win', 'glad', 'wonderful'])
        self.sad_words   = set(['sad', 'cry', 'miss', 'bad', 'sorry', 'lost', 'grief', 'pain', 'hurt', 'died', 'worst', 'fail'])
        self.fear_words  = set(['scared', 'afraid', 'fear', 'terrified', 'horror', 'danger', 'worry', 'panic', 'nervous', 'anxiety'])

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        features = []
        for text in X:
            text = str(text)
            text_lower = text.lower()
            tokens = set(re.findall(r"\w+", text_lower))

            # Structural features
            text_len   = len(text)
            caps_count = sum(1 for c in text if c.isupper())
            caps_ratio = caps_count / (text_len + 1)
            excl_count = text.count("!")
            ques_count = text.count("?")

            # Lexicon overlaps
            n_anger = len(tokens.intersection(self.anger_words))
            n_joy   = len(tokens.intersection(self.joy_words))
            n_sad   = len(tokens.intersection(self.sad_words))
            n_fear  = len(tokens.intersection(self.fear_words))

            # Simple emoticons / keywords
            has_smile = 1 if (":)" in text or ":-)" in text or "smile" in text_lower) else 0
            has_frown = 1 if (":(" in text or ":-(" in text or "sad" in text_lower) else 0

            features.append([
                text_len, caps_ratio, excl_count, ques_count,
                n_anger, n_joy, n_sad, n_fear,
                has_smile, has_frown
            ])

        return np.array(features)


# -------------------------------------------------------------------
# 3. Build train/valid/test texts (with test aligned to samplesubmission order)
# -------------------------------------------------------------------
raw_train_text = train_df["text"].apply(clean_text_final).values
y = train_df[LABEL_COL].values

# VERY IMPORTANT: align test order with sample_sub_df ids
test_for_pred = sample_sub_df[[ID_COL]].merge(
    test_df[[ID_COL, "text"]],
    on=ID_COL,
    how="left",
)
raw_test_text = test_for_pred["text"].apply(clean_text_final).values

# Hold-out split for evaluation
X_train_raw, X_valid_raw, y_train, y_valid = train_test_split(
    raw_train_text,
    y,
    test_size=0.1,
    random_state=42,
    stratify=y
)

print("\n--- Feature Engineering ---")

# -------------------------------------------------------------------
# 4. TF-IDF + hand-crafted features
# -------------------------------------------------------------------
print("Vectorizing text...")

tfidf_word = TfidfVectorizer(
    ngram_range=(1, 3),
    max_features=60000,
    min_df=2,
    sublinear_tf=True
)

tfidf_char = TfidfVectorizer(
    analyzer="char",
    ngram_range=(3, 5),
    max_features=40000,
    sublinear_tf=True
)

hand_extractor = Pipeline([
    ("extractor", AdvancedFeatures()),
    ("scaler", MinMaxScaler())
])

# Word TF-IDF
X_train_word = tfidf_word.fit_transform(X_train_raw)
X_valid_word = tfidf_word.transform(X_valid_raw)
X_test_word  = tfidf_word.transform(raw_test_text)

# Char TF-IDF
X_train_char = tfidf_char.fit_transform(X_train_raw)
X_valid_char = tfidf_char.transform(X_valid_raw)
X_test_char  = tfidf_char.transform(raw_test_text)

# Hand-crafted features (dense)
X_train_hand = hand_extractor.fit_transform(X_train_raw)
X_valid_hand = hand_extractor.transform(X_valid_raw)
X_test_hand  = hand_extractor.transform(raw_test_text)

# Combine for linear models (sparse + dense features stacked)
X_train_sparse = hstack([X_train_word, X_train_char, X_train_hand])
X_valid_sparse = hstack([X_valid_word, X_valid_char, X_valid_hand])
X_test_sparse  = hstack([X_test_word,  X_test_char,  X_test_hand])

print("Sparse feature shape:", X_train_sparse.shape)

# SVD for gradient boosting (dimensionality reduction)
print("Performing SVD for Gradient Boosting...")
svd = TruncatedSVD(n_components=120, random_state=42)

X_train_svd   = svd.fit_transform(hstack([X_train_word, X_train_char]))
X_valid_svd   = svd.transform(hstack([X_valid_word, X_valid_char]))
X_test_svd    = svd.transform(hstack([X_test_word,  X_test_char]))

X_train_dense = np.hstack([X_train_svd, X_train_hand])
X_valid_dense = np.hstack([X_valid_svd, X_valid_hand])
X_test_dense  = np.hstack([X_test_svd,  X_test_hand])

print("Dense feature shape:", X_train_dense.shape)

# -------------------------------------------------------------------
# 5. Train base models
# -------------------------------------------------------------------
print("\n--- Training Hybrid Ensemble Model ---")

# Model 1: LinearSVC (strong text baseline) + probability calibration
svm = CalibratedClassifierCV(
    LinearSVC(
        C=0.5,
        class_weight="balanced",   # handle class imbalance
        random_state=42,
        max_iter=4000
    ),
    method="sigmoid",
    cv=3
)
svm.fit(X_train_sparse, y_train)
print("SVM trained.")

# Model 2: Logistic Regression
lr = LogisticRegression(
    C=2.0,
    solver="liblinear",
    class_weight="balanced",
    random_state=42
)
lr.fit(X_train_sparse, y_train)
print("Logistic Regression trained.")

# Model 3: Multinomial Naive Bayes
nb = MultinomialNB(alpha=0.1)
nb.fit(X_train_sparse, y_train)
print("MultinomialNB trained.")

# Model 4: HistGradientBoosting on dense features
gbm = HistGradientBoostingClassifier(
    max_iter=200,
    learning_rate=0.08,
    max_depth=10,
    random_state=42
)
gbm.fit(X_train_dense, y_train)
print("HistGradientBoosting trained.")

# -------------------------------------------------------------------
# 6. Soft-voting ensemble on validation set
# -------------------------------------------------------------------
print("Ensembling on validation set...")

def get_weighted_pred(model, X, weight):
    return model.predict_proba(X) * weight

# Ensemble weights (you can tweak these)
w_svm = 3.0
w_lr  = 2.0
w_nb  = 1.0
w_gbm = 2.5

probs_valid = (
    get_weighted_pred(svm, X_valid_sparse, w_svm) +
    get_weighted_pred(lr,  X_valid_sparse, w_lr)  +
    get_weighted_pred(nb,  X_valid_sparse, w_nb)  +
    get_weighted_pred(gbm, X_valid_dense,  w_gbm)
)

valid_idx = np.argmax(probs_valid, axis=1)
valid_pred = svm.classes_[valid_idx]

acc = accuracy_score(y_valid, valid_pred)
f1  = f1_score(y_valid, valid_pred, average="macro")

print(f"Validation Accuracy: {acc:.4f}")
print(f"Validation F1 (macro): {f1:.4f}")

# -------------------------------------------------------------------
# 7. Retrain on full labeled data and create submission
# -------------------------------------------------------------------
print("\n--- Retraining on full data and predicting test set ---")

# Merge back train + valid features
X_full_sparse = vstack([X_train_sparse, X_valid_sparse])
X_full_dense  = np.vstack([X_train_dense,  X_valid_dense])
y_full        = np.concatenate([y_train, y_valid])

# Retrain all base models on full data
svm.fit(X_full_sparse, y_full)
lr.fit(X_full_sparse, y_full)
nb.fit(X_full_sparse, y_full)
gbm.fit(X_full_dense,  y_full)

# Predict on test set (already aligned with sample_sub_df ids)
probs_test = (
    get_weighted_pred(svm, X_test_sparse, w_svm) +
    get_weighted_pred(lr,  X_test_sparse, w_lr)  +
    get_weighted_pred(nb,  X_test_sparse, w_nb)  +
    get_weighted_pred(gbm, X_test_dense,  w_gbm)
)

test_idx  = np.argmax(probs_test, axis=1)
test_pred = svm.classes_[test_idx]

submission = sample_sub_df.copy()
submission[LABEL_COL] = test_pred
submission.to_csv("submission_hybrid.csv", index=False)
submission.head()
print("Saved submission_hybrid.csv")
```

```python
### Competition Code: Speed Optimized Linear Ensemble
import pandas as pd
import json
import re
import emoji
import numpy as np

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression, SGDClassifier, RidgeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from scipy.sparse import hstack, vstack

# -------------------------------------------------------------------
# 1. Load Data
# -------------------------------------------------------------------
print("--- Loading Data ---")
ID_COL = "id"
LABEL_COL = "emotion"
SUBSET_COL = "split"

emotion_df = pd.read_csv("emotion.csv")
data_id_df = pd.read_csv("data_identification.csv")
sample_sub_df = pd.read_csv("samplesubmission.csv")

with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

posts_df = pd.json_normalize(posts)
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"

posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

# Merge
full_df = posts_df.merge(data_id_df[[ID_COL, SUBSET_COL]], how="inner", on=ID_COL)
train_df = full_df[full_df[SUBSET_COL] == "train"].merge(
    emotion_df[[ID_COL, LABEL_COL]], how="left", on=ID_COL
)
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

print(f"Train rows: {len(train_df)}, Test rows: {len(test_df)}")

# -------------------------------------------------------------------
# 2. Fast Preprocessing (Regex only)
# -------------------------------------------------------------------
def clean_text_fast(text: str) -> str:
    text = str(text)
    # Demojize is useful and reasonably fast
    text = emoji.demojize(text)
    # Simple Regex cleaning
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    # Normalize common contractions
    text = text.replace("n't", " not").replace("'re", " are").replace("'s", " is")
    return text.strip()

print("Cleaning text...")
raw_train_text = train_df["text"].apply(clean_text_fast).values
y = train_df[LABEL_COL].values

# Align test data
test_for_pred = sample_sub_df[[ID_COL]].merge(
    test_df[[ID_COL, "text"]], on=ID_COL, how="left"
)
raw_test_text = test_for_pred["text"].apply(clean_text_fast).values

# Validation Split
X_train_raw, X_valid_raw, y_train, y_valid = train_test_split(
    raw_train_text, y, test_size=0.1, random_state=42, stratify=y
)

# -------------------------------------------------------------------
# 3. Fast Feature Extraction (Standard TF-IDF)
# -------------------------------------------------------------------
print("\n--- Vectorizing (Standard Optimized) ---")

# Word TF-IDF (1-3 grams is usually the sweet spot for speed/accuracy)
tfidf_word = TfidfVectorizer(
    ngram_range=(1, 3),
    max_features=60000,
    min_df=2,
    sublinear_tf=True
    # removed 'tokenizer' parameter to use fast C++ default
)

# Char TF-IDF (3-5 grams)
tfidf_char = TfidfVectorizer(
    analyzer="char",
    ngram_range=(3, 5),
    max_features=40000,
    sublinear_tf=True
)

print("1. Word TF-IDF...")
X_train_word = tfidf_word.fit_transform(X_train_raw)
X_valid_word = tfidf_word.transform(X_valid_raw)
X_test_word = tfidf_word.transform(raw_test_text)

print("2. Char TF-IDF...")
X_train_char = tfidf_char.fit_transform(X_train_raw)
X_valid_char = tfidf_char.transform(X_valid_raw)
X_test_char = tfidf_char.transform(raw_test_text)

# Stack features
X_train = hstack([X_train_word, X_train_char])
X_valid = hstack([X_valid_word, X_valid_char])
X_test = hstack([X_test_word, X_test_char])

print(f"Feature Shape: {X_train.shape}")

# -------------------------------------------------------------------
# 4. Training Fast Models
# -------------------------------------------------------------------
print("\n--- Training Fast Ensemble ---")

# Model 1: Logistic Regression (Solver 'saga' or 'liblinear' are fast)
print("Training LR...")
lr = LogisticRegression(
    C=3.0,
    solver='liblinear',
    class_weight=None, # Optimize for Accuracy
    random_state=42,
    max_iter=1000
)
lr.fit(X_train, y_train)

# Model 2: SGD Classifier (The "Fast SVM")
# loss='log_loss' gives us probabilities directly (like LR), very fast
print("Training SGD...")
sgd = SGDClassifier(
    loss='log_loss',
    alpha=1e-5,
    class_weight=None,
    random_state=42,
    max_iter=1000,
    n_jobs=-1
)
sgd.fit(X_train, y_train)

# Model 3: Multinomial NB (Blazing fast baseline)
print("Training NB...")
nb = MultinomialNB(alpha=0.1)
nb.fit(X_train, y_train)

# -------------------------------------------------------------------
# 5. Soft Voting Ensemble
# -------------------------------------------------------------------
print("\n--- Validating ---")

# Simple weighted averaging of probabilities
# LR and SGD are usually the strongest
w_lr = 2.0
w_sgd = 2.0
w_nb = 1.0

probs = (
    (lr.predict_proba(X_valid) * w_lr) +
    (sgd.predict_proba(X_valid) * w_sgd) +
    (nb.predict_proba(X_valid) * w_nb)
)

preds_idx = np.argmax(probs, axis=1)
preds = lr.classes_[preds_idx]

acc = accuracy_score(y_valid, preds)
print(f"Validation Accuracy: {acc:.4f}")

# -------------------------------------------------------------------
# 6. Retrain & Submit
# -------------------------------------------------------------------
if acc >= 0.60: # Threshold just to ensure code works
    print(">>> Retraining on full data...")

    # Merge
    X_full = vstack([X_train, X_valid])
    y_full = np.concatenate([y_train, y_valid])

    # Retrain
    lr.fit(X_full, y_full)
    sgd.fit(X_full, y_full)
    nb.fit(X_full, y_full)

    # Predict
    final_probs = (
        (lr.predict_proba(X_test) * w_lr) +
        (sgd.predict_proba(X_test) * w_sgd) +
        (nb.predict_proba(X_test) * w_nb)
    )

    final_idx = np.argmax(final_probs, axis=1)
    final_preds = lr.classes_[final_idx]

    submission = sample_sub_df.copy()
    submission[LABEL_COL] = final_preds
    submission.to_csv("submission_fast.csv", index=False)
    print("Saved submission_fast.csv")
else:
    print("Accuracy too low.")
```

```python
# 1. ÂÆâË£ùÂøÖË¶ÅÁöÑÂ•ó‰ª∂ (Â¶ÇÊûú‰Ω†Âú® ColabÔºåË´ãÁõ¥Êé•Âü∑Ë°åÈÄôË°å)
!pip install simpletransformers pandas tqdm scikit-learn

import pandas as pd
import json
import logging
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import sklearn

# Ë®≠ÂÆö LoggingÔºåËÆì‰Ω†ÁúãÂà∞Ë®ìÁ∑¥ÈÄ≤Â∫¶
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

# -------------------------------------------------------------------
# 2. Ë≥áÊñôËÆÄÂèñËàáÊï¥ÁêÜ (Data Loading)
# -------------------------------------------------------------------
print("--- Loading Data ---")
ID_COL = "id"
LABEL_COL = "emotion"
SUBSET_COL = "split"

# ËÆÄÂèñ CSV
emotion_df = pd.read_csv("emotion.csv")
data_id_df = pd.read_csv("data_identification.csv")
sample_sub_df = pd.read_csv("samplesubmission.csv")

# ËÆÄÂèñ JSON
with open("final_posts.json", "r", encoding="utf-8") as f:
    posts = json.load(f)

# Â±ïÂπ≥ JSON
posts_df = pd.json_normalize(posts)
TEXT_COL = "root._source.post.text"
POST_ID_COL = "root._source.post.post_id"

posts_df = posts_df[[POST_ID_COL, TEXT_COL]].copy()
posts_df.rename(columns={POST_ID_COL: ID_COL, TEXT_COL: "text"}, inplace=True)

# Âêà‰ΩµÊâÄÊúâË≥áÊñô
full_df = posts_df.merge(data_id_df[[ID_COL, SUBSET_COL]], how="inner", on=ID_COL)

# ÂàÜÈõ¢ Train / Test
train_df = full_df[full_df[SUBSET_COL] == "train"].merge(
    emotion_df[[ID_COL, LABEL_COL]], how="left", on=ID_COL
)
test_df = full_df[full_df[SUBSET_COL] == "test"].copy()

# Â∞çÈΩä Test È†ÜÂ∫è (Ê†πÊìö submission Ê™î)
test_df = sample_sub_df[[ID_COL]].merge(test_df[[ID_COL, "text"]], on=ID_COL, how="left")

print(f"Train rows: {len(train_df)}, Test rows: {len(test_df)}")

# -------------------------------------------------------------------
# 3. Ê†ºÂºèËΩâÊèõ (Format for SimpleTransformers)
# -------------------------------------------------------------------
# SimpleTransformers ÈúÄË¶Å DataFrame Êúâ ["text", "labels"] ÈÄôÂÖ©Ê¨Ñ
# ‰∏î labels ÂøÖÈ†àÊòØÊï∏Â≠ó (0, 1, 2...)

# Âª∫Á´ã Label Encoder
unique_labels = train_df[LABEL_COL].unique()
label_map = {label: i for i, label in enumerate(unique_labels)}
inverse_label_map = {i: label for label, i in label_map.items()}

print("Label Mapping:", label_map)

# ËΩâÊèõ Train Data
train_df['labels'] = train_df[LABEL_COL].map(label_map)
train_data = train_df[['text', 'labels']]

# ÂàáÂàÜÈ©óË≠âÈõÜ (Train: 90%, Eval: 10%)
train_split, eval_split = train_test_split(
    train_data,
    test_size=0.1,
    random_state=42,
    stratify=train_data['labels']
)

# -------------------------------------------------------------------
# 4. Ê®°ÂûãË®≠ÂÆö (RoBERTa Configuration)
# -------------------------------------------------------------------
# ÈÄô‰∫õÂèÉÊï∏ÊòØÈáùÂ∞çÈÄôÈ°ûÁ´∂Ë≥ΩÂÑ™ÂåñÈÅéÁöÑ
model_args = ClassificationArgs()
model_args.num_train_epochs = 3           # Ë∑ë 3 Ëº™ (Â§™Â∞ëÊ≤íÂ≠∏Âà∞ÔºåÂ§™Â§öÊúÉ Overfit)
model_args.learning_rate = 2e-5           # ÂæÆË™øÁöÑÊ®ôÊ∫ñÂ≠∏ÁøíÁéá
model_args.train_batch_size = 32          # Ë¶ñÈ°ØÂç°Ë®òÊÜ∂È´îË™øÊï¥ (16 Êàñ 32)
model_args.eval_batch_size = 32
model_args.max_seq_length = 128           # Êé®ÁâπÈÄöÂ∏∏Áü≠Ôºå128 Â§†‰∫Ü (ÊúÄÈï∑ÂèØË®≠ 512)
model_args.overwrite_output_dir = True
model_args.save_steps = -1                # ‰∏çÂ≠ò‰∏≠ÈñìÊ™îÔºåÁúÅÁ°¨Á¢üÁ©∫Èñì
model_args.save_model_every_epoch = False
model_args.evaluate_during_training = True
model_args.evaluate_during_training_verbose = True
model_args.use_multiprocessing = False    # ÈÅøÂÖçÊüê‰∫õÁí∞Â¢ÉÂ†±ÈåØ
model_args.use_multiprocessing_for_evaluation = False

# ÂàùÂßãÂåñÊ®°Âûã (‰ΩøÁî® roberta-base)
# RoBERTa Âú®Êé®ÁâπÊÉÖÁ∑íÂàÜÊûê‰∏äÈÄöÂ∏∏ÊØî BERT Âº∑
model = ClassificationModel(
    "roberta",
    "roberta-base",
    num_labels=len(unique_labels),
    args=model_args,
    use_cuda=True  # ÂãôÂøÖÈñãÂïü GPU
)

# -------------------------------------------------------------------
# 5. ÈñãÂßãË®ìÁ∑¥ (Training)
# -------------------------------------------------------------------
print("\n--- Starting Training (RoBERTa) ---")
# Ë®ìÁ∑¥Ê®°Âûã
model.train_model(train_split, eval_data=eval_split, acc=sklearn.metrics.accuracy_score)

# -------------------------------------------------------------------
# 6. È©óË≠âËàáË©ï‰º∞ (Evaluation)
# -------------------------------------------------------------------
print("\n--- Evaluating ---")
result, model_outputs, wrong_predictions = model.eval_model(eval_split, acc=sklearn.metrics.accuracy_score)
print(f"Validation Accuracy: {result['acc']:.4f}")

# -------------------------------------------------------------------
# 7. È†êÊ∏¨ËàáÊèê‰∫§ (Prediction)
# -------------------------------------------------------------------
print("\n--- Predicting Test Set ---")
to_predict = test_df['text'].tolist()

# ÈÄ≤Ë°åÈ†êÊ∏¨
predictions, raw_outputs = model.predict(to_predict)

# Â∞áÊï∏Â≠óËΩâÂõûÊñáÂ≠óÊ®ôÁ±§ (0 -> anger)
final_preds_labels = [inverse_label_map[pred] for pred in predictions]

# Â≠òÊ™î
submission = sample_sub_df.copy()
submission[LABEL_COL] = final_preds_labels
submission.to_csv("submission_roberta_simple.csv", index=False)
print("Saved submission_roberta_simple.csv")
```
